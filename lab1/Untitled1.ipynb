```python
import math
N = 70
p = 7
L = 3
factors = []
for a in range(2,N):
    if math.gcd(a,p) == 1:
        print('')
        print ('a = ', a)
        A = a
        for i in range(1,L+1):
            A = (A**i) % N
            F = math.gcd(A-1,N)
            if F > 1 and F < N:
                print ('F = ', F)
                if F not in factors:
                    factors.append(F)
            if F == N:
                print ('change a')
        print ('a^L! mod N = ', (a**(math.factorial(L))) %N )
print (factors)
```
